```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,1000)
h = 0.05
```

```python
fun = lambda x: (1+np.exp(-x**2))**(-1/2)
f= fun(x)
derivada_real= lambda x: (x*np.exp(-x**2))/(1+np.exp(-x**2))**(3/2)
r= derivada_real(x)
```

```python
def DerivativeC(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
resultado= DerivativeC(x,fun,h)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(2,2,1)
ax1= fig.add_subplot(2,2,2)
ax2= fig.add_subplot(2,2,3)

ax.set_title("FunciÃ³n f(x)")
ax.plot(x,f)
ax1.set_title("Derivada f'(x)")
ax1.plot(x,resultado)
ax2.set_title("Error estimado en cada punto nodal")
ax2.scatter(x,np.abs(r-resultado), s=0.5)
```
