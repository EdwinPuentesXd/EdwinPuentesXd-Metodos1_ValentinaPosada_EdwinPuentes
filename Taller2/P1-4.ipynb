```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-10,10,h)
h = 0.05
fun = lambda x: (1+np.exp(-x**2))**(-1/2)
```

$$D^{2}f(x_n)=\frac{1}{4h^{2}}\sum_{m=-2}^{2}{M[m + 2]f(x_{n−m}).}$$

```python
def convolusion_segunda_derivada(f,X):
    K=np.array([1,0,-2,0,1])
    Y= np.zeros(len(X))
    for i in range(2,len(X)-2):
        F= np.array([f(X[i-2]),f(X[i-1]),f(X[i]),f(X[i+1]),f(X[i+2])])
        Y[i]= np.dot(F,K)/(4*h**2)
    return Y
derivada2=convolusion_segunda_derivada(fun,X)
```

```python
fig=plt.figure(figsize=(7,3))
ax=fig.add_subplot(1,1,1)
ax.set_title("Segunda Derivada f''(x)")
ax.plot(X,derivada2)
```
