```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-10,10,h)
h = 0.05
fun = lambda x: (1+np.exp(-x**2))**(-1/2)
```

```python
def convolusion_derivada(f,X):
    M=np.array([-1, 0, 1])
    Y= np.zeros(len(X))
    for i in range(1,len(X)-1):
        F= np.array([f(X[i-1]),f(X[i]),f(X[i+1])])
        Y[i]= np.dot(F,M)/(2*h)
    return Y
derivada1=convolusion_derivada(fun,X)
```

```python
fig=plt.figure(figsize=(7,3))
ax=fig.add_subplot(1,1,1)
ax.set_title("Primera Derivada f'(x)")
ax.plot(X,derivada1)
```
