```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

$$Punto 10 - numeral A$$

```python
def GetT(xi,x,y):
    Sum=0
    for i in range(2):
        for j in range(2):
            Sum+=xi[i][j]*x**i*y**j
    return Sum
```

```python
a= sym.Symbol("a00")
aa= sym.Symbol("a10")
aaa= sym.Symbol("a01")
aaaa= sym.Symbol("a11")
X= sym.Symbol("x")
Y= sym.Symbol("y")

xi= [[a,aaa],[aa,aaaa]]

print("Función que define la interpolación de la temperatura del espejo: ")
GetT(xi,X,Y).expand()
```

$$Punto 10 - numeral B$$

```python
position=np.zeros((4,2))
position[0]= [1,1]
position[1]= [-1,1]
position[2]= [-1,-1]
position[3]= [1,-1]

position
```

$$Punto 10 - numeral C$$

```python
def GetM(posicion):
    M=[]
    for n in posicion:
        x=n[0]
        y=n[1]
        M.append([1,y,x,x*y])
    return np.array(M)

def Getxj(M,b):
    sol=np.linalg.solve(M,b)
    return [[sol[0],sol[1]],[sol[2],sol[3]]]
```

```python
b= np.array([1,2,0.5,0.3])
M= GetM(position)
xj= Getxj(M,b)

print("Matriz de coeficientes de la interpolación: "+str(xj))
```

$$Punto 10 - numeral D$$

```python
for n in range(len(position)):
    print("Temperatura en el vértice para " + "(X,Y)="+"("+str(position[n][0])+","+str(position[n][1])+")"+" es: " + str(GetT(xj,*position[n]))+" K")
```

```python
x= np.arange(-1,1.01,0.02)
y= np.arange(-1,1.01,0.02)
X,Y= np.meshgrid(x,y)
Z= GetT(xj,X,Y)

fig= plt.figure(figsize=(8,3))

ax = fig.add_subplot(1,2,1)
ax.pcolor(x,y,Z,cmap = 'coolwarm')
ax.scatter(np.array(position)[:,0],np.array(position)[:,1], color= "k", s=50)

ax2 = fig.add_subplot(1,2,2, projection = '3d')
ax2.plot_surface(X,Y,Z,cmap = 'coolwarm')
ax2.scatter(np.array(position)[:,0],np.array(position)[:,1],b, color= "k")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("T(x,y)")
plt.show()
```

$$Punto 10 - numeral G$$

```python
def Rotar(theta, M):
    matriz=[[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    return np.matmul(matriz,M.T).T

print("Matriz rotada 180°:")
Rotar(np.pi,position)
```

$$Punto 10 - numeral H$$

```python
def GetTM(theta,posicion=position,b=b):
    girado= Rotar(theta,posicion)
    M= GetM(girado)
    return GetT(Getxj(M,b),0,0.5)
```

$$Punto 10 - numeral I$$

```python
theta= np.linspace(0,2*np.pi,200)
menor=theta[0]
temperatura= GetTM(menor)
for n in range(1,len(theta)):
    temp= GetTM(theta[n])
    if temp<temperatura:
        temperatura=temp
        menor= theta[n]
        
menor= round(menor,3)
temperatura= round(temperatura,3)
print("La temperatura en el sensor dañado es mínima en theta="+str(menor) +" rad ("+str(round(np.degrees(menor),3))+"°)"+" y esta correspondería a "+str(temperatura)+" K.")
```
