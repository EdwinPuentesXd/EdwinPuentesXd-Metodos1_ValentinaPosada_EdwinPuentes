```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def correlacion_k_vecinos(N=10000,k=30):
    x=np.random.uniform(size=N)
    lista=np.array([])
    for k_i in range(1,k+1):
        Sum_=0
        for n in range(N-30):
            Sum_+=x[n]*x[n+k_i]
        lista= np.append(lista,Sum_)
    return lista/N
```

```python
plt.figure()
plt.plot(correlacion_k_vecinos())
plt.grid()
plt.xlim(-2,31)
plt.ylim(0.2,0.3)
plt.title("Generador Numpy")
plt.xlabel("k-esimo vecino")
plt.ylabel("C(k)")
plt.show()
```
