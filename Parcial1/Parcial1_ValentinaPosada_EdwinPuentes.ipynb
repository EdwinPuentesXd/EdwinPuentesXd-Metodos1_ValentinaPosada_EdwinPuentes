```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0.1,1.1,0.01)
tan= lambda x: (np.tan(x))**0.5
dertan= lambda x: 1/(2*(np.tan(x))**0.5*np.cos(x)**2)
y=dertan(x)
```

```python
def discretizacion(x,f,h=0.01):
    return (-3*f(x) + 4*f(x+h) - f(x+2*h))/(2*h)
progresiva= discretizacion(x,tan)
```

```python
def DerivativeC(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
central=DerivativeC(x,tan,0.01)
```

```python
fig= plt.figure(figsize=(9,7))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.set_title("Derivada Central")
ax1.plot(x,central, "r*")
ax1.plot(x,y, "k")

ax2.set_title("Derivada Progresiva")
ax2.plot(x,progresiva, "g*")
ax2.plot(x,y, "k")

ax3.set_title("Error Derivada Central")
ax3.scatter(x,np.abs(y-central), color="r",s=0.5)
ax4.set_title("Error Derivada Progresiva")
ax4.scatter(x,np.abs(y-progresiva), color="g", s=0.5)
```

```python
print("Valor mÃ¡ximo de error en derivada progresiva: " + str(np.max(np.abs(y-progresiva))))
print("Valor mÃ¡ximo de error en derivada central: " + str(np.max(np.abs(y-central))))
```

$$Conclusiones$$

Se puede observar que tanto la derivada progresiva como la derivada central tienen una muy buena aproximaciÃ³n respecto a la derivada real, al punto que si se grafican sobre la derivada real son casi indistinguibles. Respecto al error en los puntos nodales podemos observar que la derivada central es un poco mÃ¡s precisa que la derivada progresiva debido que el error mÃ¡ximo de la derivada progresiva es 0.003 mientras que el de la central es 0.002. Sin embargo, consideramos que la derivada progresiva calculada por el polinomio de interpolaciÃ³n es bastante parecida a la derivada central y a la derivada real, por tanto es aceptable utilizar este polinomio de interpolaciÃ³n para estimar la derivada de una funciÃ³n.
