b) Un diferencial de la flor se puede considerar como un $\Delta A$, dado por, 

$$
\Delta A = \frac{\Delta \theta}{2}\Big[\frac{r_i+r_{i+1}}{2}\Big]^2 
$$

Considerando ese diferencial de area, imprima el area de la flor tomando 500 pasos para el ángulo $\theta$ 



````python
def dA(x,flor):
    area=0
    for n in range(len(x)-1):
        area+= ((flor(x[n])+flor(x[n+1]))/2)**2
    return area*dx/2

da= dA(x,flor)
da
````
