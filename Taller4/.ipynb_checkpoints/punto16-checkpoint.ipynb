```python
import matplotlib.pyplot as plt
import numpy as np
```

*Parte a*

```python
funcion=lambda x : np.exp(x)*x**3/(np.exp(x)-1)
real=np.pi**4/15

def laguerre(f, n):
    Roots,Weights=np.polynomial.laguerre.laggauss(n)
    return np.sum(Weights*funcion(Roots))

resultado=laguerre(funcion, 3)
resultado
```

*Parte b*

```python
y=np.zeros(0)
x=np.arange(2,11,1)
for n in x:
    integral=laguerre(funcion,n)
    error=(integral/real)
    y=np.append(y,error)
    
plt.figure(figsize=(5,5))
plt.plot(x,y,"bo")
plt.grid()
plt.ylabel("Er(n)")
plt.xlabel("n")
plt.legend(["Laguerre quadrature acurracy"], loc="lower right")
plt.show()
```
